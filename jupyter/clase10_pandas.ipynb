# 3.2. Imputar Valores Perdidos

```python
import numpy as np
```

```python
# Crear un data frame en base a números aleatorios.
df = pd.DataFrame(np.random.randn(5, 3), index=['a0', 'a10','a20', 'a30', 'a40'],columns=['X', 'Y', 'Z'])
df
```

```python
# Crear índices adicionales al data frame.
df2 = df.reindex(['a0', 'a1', 'a10', 'a11', 'a20', 'a21','a30', 'a31', 'a40', 'a41'])
df2
```

```python
# Completar los valores perdidos con ceros.
df3 = df2.fillna(0)
df3
```

```python
# Completar los valores perdidos con el método “forward propagation”. Se va completar con el valor previo al nulo.
df4 = df2.fillna(method='pad')
df4
```

```python
# Completar los valores perdidos con el promedio de la variable.
df5 = df2.fillna(df2.mean())
df5
```

```python
df22 = df.reindex(['a99','a0', 'a1', 'a10', 'a11', 'a20', 'a21','a30', 'a31', 'a40', 'a41'])
df22
```

```python
df22.fillna(method='pad')
```

```python
from sklearn.preprocessing import Imputer
imp = Imputer(missing_values="NaN", strategy='median', axis=0)
df23 = imp.fit_transform(df22)
```

```python
df23
```

```python
df24 = pd.DataFrame(df23)
```

```python
# Crear datos
grade_lookup = {'GRADE': pd.Series(['ELEMENTARY','MIDDLE/HIGH', 'MISC']),'LEVEL': pd.Series([1, 2, 3])}
grade_lookup
```

```python
# Primeros 5 elementos del campo GRADE LEVEL
df[['GRADE LEVEL']][0:7]
```

```python
# Inner Join
d_sub = df[0:5].join(grade_lookup2.set_index(['GRADE']),on=['GRADE LEVEL'], how='inner')
d_sub
```

```python
#Left Join.
d_sub = df[0:5].join(grade_lookup2.set_index(['GRADE']),on=['GRADE LEVEL'], how='left')
d_sub
```

```python
#Full Outer Join.
d_sub = df[0:5].join(grade_lookup2.set_index(['GRADE']),on=['GRADE LEVEL'], how='outer')
d_sub
```
